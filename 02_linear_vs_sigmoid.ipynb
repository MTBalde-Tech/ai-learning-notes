## 1. Linear Function

A linear function has the form:

\[
y = wx + b
\]

Let's define and plot a simple linear function.


```python

import numpy as np
import matplotlib.pyplot as plt

def linear(x, w=1.0, b=0.0):
    return w * x + b

x = np.linspace(-10, 10, 400)
y_linear = linear(x, w=0.8, b=0.0)

plt.figure()
plt.plot(x, y_linear)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Function: y = 0.8x")
plt.grid(True)
plt.show()
```
